### Problem 10

The variables `q10` is initialized with 5 steps of the Collatz sequence starting from 15.
Collatz sequences work with the following rule.

```
if the current value x is odd:
    the next value will be 3x+1
else:
    the next value will be x/2
```

Extend `q10` to ten entries following this rule, then plot them as a line chart.
Use the values in `q10` as the y values.
You can use the default values for x in the `plt.plot` call.

````python
# DO NOT CHANGE

q10 = [15, 46, 23, 70, 35]
````

````python
# YOUR CHANGES HERE

# Extend q10 to ten entries
while len(q10) < 10:
    current_value = q10[-1]
    if current_value % 2 == 0:  # If even
        next_value = current_value // 2
    else:  # If odd
        next_value = 3 * current_value + 1
    q10.append(next_value)

# Plot the extended q10 as a line chart
plt.plot(q10)
plt.show()
````

### Problem 15

Use the function `math.comb` to make a bar chart illustrating the binomial distribution for $n=10$.
That is, for integers x from zero to ten, there should be a bar whose height is $10 \choose x$.

````python
# YOUR CHANGES HERE

# The number of trials
n = 10

# Create a list of integers from 0 to 10 for the x-axis
x_values = list(range(n + 1))

# Calculate the binomial coefficient (n choose k) for each value of k
# Here, `k` is the variable `x` from the problem statement
y_values = [math.comb(n, k) for k in x_values]

# Create the bar chart
plt.bar(x_values, y_values)

# Add labels and a title for clarity
plt.title("Binomial Distribution")

# Set the x-axis ticks to be the integers from 0 to 10
plt.xticks(x_values)

# Display the plot
plt.show()
````

### Problem 17

Plot the factorial function from n=1 to 10 using log scale for better legibility.

Recall that the factorial of $n$, denoted $n!$, is calculated as the product $1 \times 2 \times \ldots \times n$.
The plots that you plot should be $(1, 1!), (2, 2!), (3, 3!), \ldots, (10, 10!)$.
Use [`plt.yscale`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yscale.html) to change your y axis to use log scale.

````python
# YOUR CHANGES HERE

# Define the range of n for the x-axis
x_values = range(1, 11)

# Calculate the factorial for each value of n
y_values = [math.factorial(n) for n in x_values]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the points
plt.plot(x_values, y_values)

# Set the y-axis to a logarithmic scale.
plt.yscale("log")

# Add labels and a title for clarity.
plt.xlabel('n')
plt.ylabel('n! (Factorial)')
plt.title('Factorial Function on a Logarithmic Scale')

# Display the plot.
plt.show()
````

### Problem 18

Read the data from `f18.tsv` and parse it as a TSV file.
Divide the data by the "animal" column and in a subplot for each animal, plot the "age" and "weight" columns as the x and y axes respectively with a different subplot for each animal.
Each subplot should be a line chart, but add markers for each point as some animals only have one data point.
Make sure to mention the animal name in the chart titles, and label the axes.

````python
# YOUR CHANGES HERE

# Import
import pandas as pd

# Read the TSV file into a pandas DataFrame
df = pd.read_csv("f18.tsv", sep="\t")

# Get unique animal names
animals = df["animal"].unique()

# Create a figure and a set of subplots
# Adjust figsize based on the number of unique animals for better visualization
fig, axes = plt.subplots(len(animals), 1, figsize=(8, 4 * len(animals)))

# If there's only one animal, axes will not be an array, so convert it for consistent indexing
if len(animals) == 1:
    axes = [axes]

# Iterate through each animal and create a subplot
for i, animal in enumerate(animals):
    animal_data = df[df["animal"] == animal]

    # Plot age vs. weight for the current animal
    axes[i].plot(animal_data["age"], animal_data["weight"], marker="*")

    # Set title and labels for the subplot
    axes[i].set_title(f"{animal} - Age vs. Weight")
    axes[i].set_xlabel("Age")
    axes[i].set_ylabel("Weight")

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()
plt.show()
````

````python
# YOUR CHANGES HERE

#Import
import pandas as pd

# Read in the file
df = pd.read_csv("f20.tsv", sep="\t")

# Create a scatter plot using the x, y, and color columns.
# df["x"] retrieves the data from the x column.
# df["y"] retrieves the data from the "y" column.
# df["color"] retrieves the data from the color column, which matplotlib interprets to color each point accordingly.
plt.scatter(df["x"], df["y"], c = df["color"])

# Add a title and labels for clarity (optional).
plt.title("Scatter Plot of x vs y from f20.tsv")
plt.xlabel("x")
plt.ylabel("y")

# Display the plot.
plt.show()
````
